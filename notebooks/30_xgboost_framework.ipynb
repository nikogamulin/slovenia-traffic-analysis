# Story 2.1: XGBoost Training Framework

This notebook implements a reusable XGBoost training framework for sensor-level predictions.

## Features:
- Base training pipeline for XGBoost models
- Hyperparameter tuning capabilities
- Cross-validation strategies
- Model evaluation utilities
- Support for both regression and classification tasks

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split, cross_val_score, TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
from pathlib import Path
from datetime import datetime
from typing import Dict, List, Tuple, Optional, Any
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette('husl')

print("XGBoost version:", xgb.__version__)
print("Framework initialized successfully!")
```

## 1. Generate Synthetic Training Data

```python
def generate_synthetic_sensor_data(n_samples: int = 5000, n_sensors: int = 5, seed: int = 42) -> pd.DataFrame:
    """Generate synthetic sensor data for demonstration."""
    np.random.seed(seed)
    
    data_list = []
    
    for sensor_id in range(n_sensors):
        # Generate time series
        timestamps = pd.date_range(start='2023-01-01', periods=n_samples//n_sensors, freq='H')
        
        # Generate features
        sensor_data = pd.DataFrame({
            'timestamp': timestamps,
            'sensor_id': f'sensor_{sensor_id:03d}',
            
            # Traffic features
            'hour': timestamps.hour,
            'day_of_week': timestamps.dayofweek,
            'month': timestamps.month,
            'is_weekend': (timestamps.dayofweek >= 5).astype(int),
            
            # Historical features (lagged values)
            'traffic_lag_1h': np.random.poisson(100, len(timestamps)) + np.random.normal(0, 10, len(timestamps)),
            'traffic_lag_24h': np.random.poisson(100, len(timestamps)) + np.random.normal(0, 15, len(timestamps)),
            'traffic_lag_168h': np.random.poisson(100, len(timestamps)) + np.random.normal(0, 20, len(timestamps)),
            
            'speed_lag_1h': np.clip(np.random.normal(80, 15, len(timestamps)), 20, 130),
            'speed_lag_24h': np.clip(np.random.normal(80, 15, len(timestamps)), 20, 130),
            
            # Weather features
            'temperature': np.random.normal(15, 10, len(timestamps)),
            'precipitation': np.clip(np.random.exponential(2, len(timestamps)), 0, 50),
            'visibility': np.clip(np.random.normal(10, 3, len(timestamps)), 0.1, 15),
            'wind_speed': np.clip(np.random.exponential(3, len(timestamps)), 0, 30),
            
            # Calendar features
            'is_holiday': np.random.choice([0, 1], len(timestamps), p=[0.95, 0.05]),
            'is_school_break': np.random.choice([0, 1], len(timestamps), p=[0.85, 0.15]),
            
            # Incident features
            'has_accident': np.random.choice([0, 1], len(timestamps), p=[0.98, 0.02]),
            'has_roadwork': np.random.choice([0, 1], len(timestamps), p=[0.95, 0.05]),
        })
        
        # Generate targets with some realistic patterns
        base_traffic = 100 + sensor_id * 20
        base_speed = 80 - sensor_id * 5
        
        # Traffic volume target
        sensor_data['vehicle_count'] = (
            base_traffic +
            30 * np.sin(2 * np.pi * sensor_data['hour'] / 24) +  # Daily pattern
            20 * (1 - sensor_data['is_weekend']) +  # Weekday effect
            -10 * sensor_data['precipitation'] / 10 +  # Weather effect
            -50 * sensor_data['has_accident'] +  # Accident effect
            np.random.normal(0, 10, len(timestamps))  # Noise
        )
        
        # Average speed target
        sensor_data['avg_speed'] = np.clip(
            base_speed +
            -10 * np.sin(2 * np.pi * sensor_data['hour'] / 24) +  # Daily pattern
            5 * sensor_data['is_weekend'] +  # Weekend effect
            -2 * sensor_data['precipitation'] +  # Weather effect
            -30 * sensor_data['has_accident'] +  # Accident effect
            np.random.normal(0, 5, len(timestamps)),  # Noise
            20, 130
        )
        
        data_list.append(sensor_data)
    
    return pd.concat(data_list, ignore_index=True)

# Generate data
synthetic_data = generate_synthetic_sensor_data(n_samples=5000, n_sensors=5)
print(f"Generated {len(synthetic_data)} samples for {synthetic_data['sensor_id'].nunique()} sensors")
print(f"Data shape: {synthetic_data.shape}")
print(f"\nSensor IDs: {synthetic_data['sensor_id'].unique().tolist()}")
print(f"\nFeature columns: {[col for col in synthetic_data.columns if col not in ['timestamp', 'sensor_id', 'vehicle_count', 'avg_speed']]}")
print(f"\nTarget columns: ['vehicle_count', 'avg_speed']")

# Display sample data
synthetic_data.head()
```

## 2. XGBoost Training Framework Class

```python
class XGBoostFramework:
    """Reusable XGBoost training framework for sensor-level predictions."""
    
    def __init__(self, 
                 task_type: str = 'regression',
                 model_name: str = 'xgboost_model',
                 save_dir: str = './models'):
        """
        Initialize XGBoost Framework.
        
        Args:
            task_type: 'regression' or 'classification'
            model_name: Name for the model
            save_dir: Directory to save models
        """
        self.task_type = task_type
        self.model_name = model_name
        self.save_dir = Path(save_dir)
        self.save_dir.mkdir(exist_ok=True)
        
        self.model = None
        self.best_params = None
        self.feature_importance = None
        self.training_history = []
        self.scaler = StandardScaler()
        
    def get_default_params(self) -> Dict[str, Any]:
        """Get default XGBoost parameters based on task type."""
        base_params = {
            'n_estimators': 100,
            'max_depth': 6,
            'learning_rate': 0.1,
            'subsample': 0.8,
            'colsample_bytree': 0.8,
            'min_child_weight': 1,
            'gamma': 0,
            'reg_alpha': 0,
            'reg_lambda': 1,
            'random_state': 42,
            'n_jobs': -1,
            'verbosity': 0
        }
        
        if self.task_type == 'regression':
            base_params['objective'] = 'reg:squarederror'
            base_params['eval_metric'] = 'rmse'
        else:
            base_params['objective'] = 'binary:logistic'
            base_params['eval_metric'] = 'logloss'
        
        return base_params
    
    def prepare_features(self, 
                        df: pd.DataFrame, 
                        target_col: str,
                        feature_cols: List[str] = None,
                        fit_scaler: bool = True) -> Tuple[np.ndarray, np.ndarray]:
        """Prepare features and target for training."""
        
        if feature_cols is None:
            # Use all columns except target, timestamp, and sensor_id
            exclude_cols = [target_col, 'timestamp', 'sensor_id']
            feature_cols = [col for col in df.columns if col not in exclude_cols]
        
        X = df[feature_cols].values
        y = df[target_col].values if target_col in df.columns else None
        
        # Scale features
        if fit_scaler:
            X = self.scaler.fit_transform(X)
        else:
            X = self.scaler.transform(X)
        
        self.feature_names = feature_cols
        
        return X, y
    
    def train(self, 
             X_train: np.ndarray, 
             y_train: np.ndarray,
             X_val: np.ndarray = None,
             y_val: np.ndarray = None,
             params: Dict[str, Any] = None,
             early_stopping_rounds: int = 10) -> xgb.XGBRegressor:
        """Train XGBoost model."""
        
        if params is None:
            params = self.get_default_params()
        
        # Add early_stopping_rounds to params if validation set is provided
        if X_val is not None and y_val is not None and early_stopping_rounds:
            params['early_stopping_rounds'] = early_stopping_rounds
        
        # Initialize model
        if self.task_type == 'regression':
            self.model = xgb.XGBRegressor(**params)
        else:
            self.model = xgb.XGBClassifier(**params)
        
        # Prepare evaluation set
        eval_set = None
        if X_val is not None and y_val is not None:
            eval_set = [(X_val, y_val)]
        
        # Train model
        self.model.fit(
            X_train, y_train,
            eval_set=eval_set,
            verbose=False
        )
        
        # Store feature importance
        self.feature_importance = pd.DataFrame({
            'feature': self.feature_names,
            'importance': self.model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        # Record training
        self.training_history.append({
            'timestamp': datetime.now().isoformat(),
            'params': params,
            'n_features': len(self.feature_names),
            'n_train_samples': len(X_train),
            'n_val_samples': len(X_val) if X_val is not None else 0
        })
        
        return self.model
    
    def evaluate(self, 
                X_test: np.ndarray, 
                y_test: np.ndarray) -> Dict[str, float]:
        """Evaluate model performance."""
        
        if self.model is None:
            raise ValueError("Model not trained yet!")
        
        y_pred = self.model.predict(X_test)
        
        if self.task_type == 'regression':
            metrics = {
                'rmse': np.sqrt(mean_squared_error(y_test, y_pred)),
                'mae': mean_absolute_error(y_test, y_pred),
                'r2': r2_score(y_test, y_pred),
                'mape': np.mean(np.abs((y_test - y_pred) / (y_test + 1e-10))) * 100
            }
        else:
            from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
            metrics = {
                'accuracy': accuracy_score(y_test, y_pred > 0.5),
                'precision': precision_score(y_test, y_pred > 0.5),
                'recall': recall_score(y_test, y_pred > 0.5),
                'f1': f1_score(y_test, y_pred > 0.5)
            }
        
        return metrics
    
    def cross_validate(self, 
                      X: np.ndarray, 
                      y: np.ndarray,
                      cv_folds: int = 5,
                      params: Dict[str, Any] = None) -> Dict[str, Any]:
        """Perform cross-validation."""
        
        if params is None:
            params = self.get_default_params()
        
        # Initialize model for CV
        if self.task_type == 'regression':
            cv_model = xgb.XGBRegressor(**params)
            scoring = 'neg_mean_squared_error'
        else:
            cv_model = xgb.XGBClassifier(**params)
            scoring = 'roc_auc'
        
        # Perform cross-validation
        cv_scores = cross_val_score(
            cv_model, X, y, 
            cv=cv_folds, 
            scoring=scoring,
            n_jobs=-1
        )
        
        if self.task_type == 'regression':
            cv_scores = np.sqrt(-cv_scores)  # Convert to RMSE
        
        return {
            'mean_score': cv_scores.mean(),
            'std_score': cv_scores.std(),
            'scores': cv_scores.tolist()
        }
    
    def time_series_cv(self, 
                      X: np.ndarray, 
                      y: np.ndarray,
                      n_splits: int = 5,
                      params: Dict[str, Any] = None) -> Dict[str, Any]:
        """Perform time series cross-validation."""
        
        if params is None:
            params = self.get_default_params()
        
        tscv = TimeSeriesSplit(n_splits=n_splits)
        scores = []
        
        for train_idx, val_idx in tscv.split(X):
            X_train_cv, X_val_cv = X[train_idx], X[val_idx]
            y_train_cv, y_val_cv = y[train_idx], y[val_idx]
            
            # Train model
            if self.task_type == 'regression':
                cv_model = xgb.XGBRegressor(**params)
            else:
                cv_model = xgb.XGBClassifier(**params)
            
            cv_model.fit(X_train_cv, y_train_cv, verbose=False)
            
            # Evaluate
            y_pred = cv_model.predict(X_val_cv)
            
            if self.task_type == 'regression':
                score = np.sqrt(mean_squared_error(y_val_cv, y_pred))
            else:
                from sklearn.metrics import roc_auc_score
                score = roc_auc_score(y_val_cv, y_pred)
            
            scores.append(score)
        
        return {
            'mean_score': np.mean(scores),
            'std_score': np.std(scores),
            'scores': scores
        }
    
    def plot_feature_importance(self, top_n: int = 20):
        """Plot feature importance."""
        
        if self.feature_importance is None:
            raise ValueError("Model not trained yet!")
        
        plt.figure(figsize=(10, 8))
        
        # Get top features
        top_features = self.feature_importance.head(top_n)
        
        # Create plot
        plt.barh(range(len(top_features)), top_features['importance'])
        plt.yticks(range(len(top_features)), top_features['feature'])
        plt.xlabel('Feature Importance')
        plt.title(f'Top {top_n} Feature Importances')
        plt.gca().invert_yaxis()
        
        plt.tight_layout()
        plt.show()
    
    def plot_predictions(self, y_true: np.ndarray, y_pred: np.ndarray, sample_size: int = 100):
        """Plot predictions vs actual values."""
        
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        
        # Sample data for visualization
        if len(y_true) > sample_size:
            idx = np.random.choice(len(y_true), sample_size, replace=False)
            y_true_sample = y_true[idx]
            y_pred_sample = y_pred[idx]
        else:
            y_true_sample = y_true
            y_pred_sample = y_pred
        
        # Scatter plot
        axes[0].scatter(y_true_sample, y_pred_sample, alpha=0.5)
        axes[0].plot([y_true_sample.min(), y_true_sample.max()], 
                    [y_true_sample.min(), y_true_sample.max()], 
                    'r--', lw=2)
        axes[0].set_xlabel('Actual')
        axes[0].set_ylabel('Predicted')
        axes[0].set_title('Predictions vs Actual')
        
        # Residual plot
        residuals = y_true_sample - y_pred_sample
        axes[1].scatter(y_pred_sample, residuals, alpha=0.5)
        axes[1].axhline(y=0, color='r', linestyle='--')
        axes[1].set_xlabel('Predicted')
        axes[1].set_ylabel('Residuals')
        axes[1].set_title('Residual Plot')
        
        plt.tight_layout()
        plt.show()
    
    def save_model(self, path: str = None):
        """Save model and metadata."""
        
        if self.model is None:
            raise ValueError("No model to save!")
        
        if path is None:
            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
            path = self.save_dir / f"{self.model_name}_{timestamp}"
        else:
            path = Path(path)
        
        path.mkdir(exist_ok=True, parents=True)
        
        # Save model
        joblib.dump(self.model, path / 'model.pkl')
        
        # Save scaler
        joblib.dump(self.scaler, path / 'scaler.pkl')
        
        # Save metadata
        metadata = {
            'model_name': self.model_name,
            'task_type': self.task_type,
            'feature_names': self.feature_names,
            'best_params': self.best_params,
            'training_history': self.training_history,
            'feature_importance': self.feature_importance.to_dict() if self.feature_importance is not None else None
        }
        
        with open(path / 'metadata.json', 'w') as f:
            json.dump(metadata, f, indent=2, default=str)
        
        print(f"Model saved to {path}")
        return str(path)
    
    def load_model(self, path: str):
        """Load model and metadata."""
        
        path = Path(path)
        
        # Load model
        self.model = joblib.load(path / 'model.pkl')
        
        # Load scaler
        self.scaler = joblib.load(path / 'scaler.pkl')
        
        # Load metadata
        with open(path / 'metadata.json', 'r') as f:
            metadata = json.load(f)
        
        self.model_name = metadata['model_name']
        self.task_type = metadata['task_type']
        self.feature_names = metadata['feature_names']
        self.best_params = metadata['best_params']
        self.training_history = metadata['training_history']
        
        if metadata['feature_importance']:
            self.feature_importance = pd.DataFrame(metadata['feature_importance'])
        
        print(f"Model loaded from {path}")

# Test the framework
framework = XGBoostFramework(task_type='regression', model_name='test_model')
print("XGBoostFramework initialized successfully!")
print(f"Default parameters: {framework.get_default_params()}")
```

## 3. Train Example Model

```python
# Prepare data for a single sensor
sensor_id = 'sensor_000'
sensor_data = synthetic_data[synthetic_data['sensor_id'] == sensor_id].copy()

print(f"Training model for {sensor_id}")
print(f"Data shape: {sensor_data.shape}")

# Initialize framework for vehicle count prediction
vehicle_count_framework = XGBoostFramework(
    task_type='regression',
    model_name=f'{sensor_id}_vehicle_count'
)

# Prepare features
X, y = vehicle_count_framework.prepare_features(
    sensor_data, 
    target_col='vehicle_count'
)

# Split data
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print(f"\nTraining set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")

# Train model
model = vehicle_count_framework.train(
    X_train, y_train,
    X_val=X_test, y_val=y_test,
    early_stopping_rounds=10
)

print("\nModel trained successfully!")

# Evaluate model
metrics = vehicle_count_framework.evaluate(X_test, y_test)
print(f"\nTest Metrics:")
for metric, value in metrics.items():
    print(f"  {metric.upper()}: {value:.3f}")

# Check if RMSE < 15% of mean
mean_vehicle_count = y_test.mean()
rmse_percentage = (metrics['rmse'] / mean_vehicle_count) * 100
print(f"\nRMSE as % of mean: {rmse_percentage:.1f}%")
print(f"Target: < 15% - {'✅ PASSED' if rmse_percentage < 15 else '❌ FAILED'}")
```

## 4. Cross-Validation Testing

```python
# Test cross-validation
print("Testing cross-validation strategies...\n")

# Standard cross-validation
cv_results = vehicle_count_framework.cross_validate(X, y, cv_folds=5)
print("Standard Cross-Validation (5-fold):")
print(f"  Mean RMSE: {cv_results['mean_score']:.3f}")
print(f"  Std RMSE: {cv_results['std_score']:.3f}")
print(f"  Fold scores: {[f'{s:.3f}' for s in cv_results['scores']]}")

# Time series cross-validation
tscv_results = vehicle_count_framework.time_series_cv(X, y, n_splits=5)
print("\nTime Series Cross-Validation (5-fold):")
print(f"  Mean RMSE: {tscv_results['mean_score']:.3f}")
print(f"  Std RMSE: {tscv_results['std_score']:.3f}")
print(f"  Fold scores: {[f'{s:.3f}' for s in tscv_results['scores']]}")

# Visualize CV results
fig, ax = plt.subplots(figsize=(10, 6))

x_labels = [f'Fold {i+1}' for i in range(5)]
x_pos = np.arange(len(x_labels))
width = 0.35

ax.bar(x_pos - width/2, cv_results['scores'], width, label='Standard CV', alpha=0.8)
ax.bar(x_pos + width/2, tscv_results['scores'], width, label='Time Series CV', alpha=0.8)

ax.set_xlabel('Fold')
ax.set_ylabel('RMSE')
ax.set_title('Cross-Validation Comparison')
ax.set_xticks(x_pos)
ax.set_xticklabels(x_labels)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Train Models for All Sensors

```python
# Train models for all sensors
sensor_models = {}
sensor_metrics = []

print("Training models for all sensors...\n")

for sensor_id in synthetic_data['sensor_id'].unique():
    print(f"Training for {sensor_id}...")
    
    # Get sensor data
    sensor_data = synthetic_data[synthetic_data['sensor_id'] == sensor_id].copy()
    
    # Initialize framework
    framework = XGBoostFramework(
        task_type='regression',
        model_name=f'{sensor_id}_vehicle_count'
    )
    
    # Prepare features
    X, y = framework.prepare_features(sensor_data, target_col='vehicle_count')
    
    # Split data
    split_idx = int(len(X) * 0.8)
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]
    
    # Train model
    model = framework.train(X_train, y_train, X_test, y_test)
    
    # Evaluate
    metrics = framework.evaluate(X_test, y_test)
    
    # Store results
    sensor_models[sensor_id] = framework
    sensor_metrics.append({
        'sensor_id': sensor_id,
        'rmse': metrics['rmse'],
        'mae': metrics['mae'],
        'r2': metrics['r2'],
        'mape': metrics['mape'],
        'mean_target': y_test.mean(),
        'rmse_percentage': (metrics['rmse'] / y_test.mean()) * 100
    })
    
    print(f"  RMSE: {metrics['rmse']:.3f} ({(metrics['rmse'] / y_test.mean()) * 100:.1f}% of mean)")
    print(f"  R2: {metrics['r2']:.3f}\n")

# Summary DataFrame
metrics_df = pd.DataFrame(sensor_metrics)
print("\n" + "="*60)
print("SUMMARY OF ALL SENSOR MODELS")
print("="*60)
print(metrics_df.to_string())

# Check success criteria
print("\n" + "="*60)
print("SUCCESS CRITERIA CHECK")
print("="*60)
success_count = (metrics_df['rmse_percentage'] < 15).sum()
total_count = len(metrics_df)
print(f"Models meeting RMSE < 15% of mean: {success_count}/{total_count}")
print(f"Success rate: {success_count/total_count*100:.1f}%")
print(f"Average RMSE %: {metrics_df['rmse_percentage'].mean():.1f}%")
```
